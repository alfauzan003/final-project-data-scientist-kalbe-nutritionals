# Data Wrangling

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

## Getting Data

```python
customers = pd.read_csv('../Data/raw/Case Study - Customer.csv', sep=';')
customers.head()
```

```python
products = pd.read_csv('../Data/raw/Case Study - Product.csv', sep=';')
products.head()
```

```python
stores = pd.read_csv('../Data/raw/Case Study - Store.csv', sep=';')
stores.head()
```

```python
transactions = pd.read_csv('../Data/raw/Case Study - Transaction.csv', sep=';')
transactions.head()
```

## Assesing Data

```python
customers.info()
print(customers.isna().sum())
print("Jumlah duplikasi: ", customers.duplicated().sum())
customers.describe()
```

```python
products.info()
print(products.isna().sum())
print("Jumlah duplikasi: ", products.duplicated().sum())
products.describe()
```

```python
stores.info()
print(stores.isna().sum())
print("Jumlah duplikasi: ", stores.duplicated().sum())
stores.describe()
```

```python
transactions.info()
print(transactions.isna().sum())
print("Jumlah duplikasi: ", transactions.duplicated().sum())
transactions.describe()
```

## Cleaning Data

```python
# Data Customer

# Mengubah tipe data 'Income' menjadi float
customers['Income'] = customers['Income'].astype(str).str.replace(',', '.').astype(float)
# Mengubah tipe data 'CustomerID' menjadi string
customers['CustomerID'] = customers['CustomerID'].astype(str)
# Menghapus baris yang memiliki nilai kosong
customers.dropna(inplace=True)
# Menghapus duplikasi
customers.drop_duplicates(inplace=True)

customers.info()
print(customers.isna().sum())
```

```python
# Data Stores
# Mengubah tipe data 'StoreID' menjadi string
stores['StoreID'] = stores['StoreID'].astype(str)
stores['Latitude'] = stores['Latitude'].astype(str).str.replace(',', '.').astype(float)
stores['Longitude'] = stores['Longitude'].astype(str).str.replace(',', '.').astype(float)

stores.info()
```

```python
# Data Transactions
# Mengubah tipe data 'CustomerID' menjadi string
transactions['CustomerID'] = transactions['CustomerID'].astype(str)
# Mengubah tipe data 'Date' menjadi datetime format dd/mm/yyyy
transactions['Date'] = pd.to_datetime(transactions['Date'], format='%d/%m/%Y')
# Mengubah tipe data 'StoreID' menjadi string
transactions['StoreID'] = transactions['StoreID'].astype(str)

transactions.info()
```

## Merge and Save Data

```python
# Merging Data
gabung1 = pd.merge(transactions, customers, how='left', on='CustomerID')
gabung2 = pd.merge(gabung1, stores, how='left', on='StoreID')
gabung3 = pd.merge(gabung2, products.drop(columns=['Price']), how='left', on='ProductID', )

merged = gabung3.copy()

merged.dropna(inplace=True)
merged['Age'] = merged['Age'].astype(int)
merged['Gender'] = merged['Gender'].astype(int)
merged.info()

merged.head()
```

```python
# Save data
customers.to_csv('../Data/clean/customers.csv', index=False)
products.to_csv('../Data/clean/products.csv', index=False)
stores.to_csv('../Data/clean/stores.csv', index=False)
transactions.to_csv('../Data/clean/transactions.csv', index=False)

merged.to_csv('../Data/clean/merged.csv', index=False)
```
